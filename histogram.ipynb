# Histogram

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Data Loading
iris = sns.load_dataset('iris')
iris.shape
```

```python
iris.head()
```

```python
iris.groupby('species').size()
```

## (1) Histogram via matplotlib.pyplot

```python
plt.hist(iris['sepal_width'], bins=10)
```

```python
n, bins, patches = plt.hist(iris['sepal_width'], bins=10)
```

```python
n, bins, patches
```

```python
# density = True
n, bins, patches = plt.hist(iris['sepal_width'], bins=10, density=True)
```

```python
# facecolor, alpha
n, bins, patches = plt.hist(iris['sepal_width'], bins=10, 
                            density=True, 
                            facecolor='blue', 
                            alpha=0.5)
```

```python
# Setting X, Y label and Title, axis
n, bins, patches = plt.hist(iris['sepal_width'], 
                            bins=10, 
                            density=True, 
                            facecolor='blue', 
                            alpha=0.5)
plt.xlabel('X bins')
plt.ylabel('Density')
plt.title('Histogram of Sepal Width')
plt.axis([1.5, 4.5, 0, 1.1])
plt.show()
```

### (2) Histogram via seaborn

```python
sns.distplot(iris['sepal_width'])
plt.show()
```

```python
# Kernel Density Curve with histogram
ax = sns.distplot(iris['sepal_width'], 
                  hist=True, 
                  kde=True, 
                  bins=10, 
                  color='blue', 
                  hist_kws={'edgecolor': 'gray'}, 
                  kde_kws={'linewidth': 2})
ax.set_title('Histogram of Sepal Width')
ax.set_xlabel('Sepal Width(cm)')
ax.set_ylabel('Density')
plt.show()
```

```python
sns.distplot(iris['sepal_width'],  
             hist=False)
plt.show()
```

```python
# Kernel Density Estimate Plot
sns.kdeplot(iris['sepal_width'],  
             shade=True, 
           bw=2, 
           label="Sepal Width")
plt.legend()
plt.show()
```

### (3) Histogram by pandas.DataFrame.hist

```python
# Fixing random state for reproducibility
np.random.seed(123)
```

```python
# Generating random numbers from gaussian distribution
mu, sigma = 0, 3
x = mu + sigma * np.random.randn(1000)
```

```python
# Histogram of x
plt.hist(x, bins=20)
```

```python
n, bins, patches = plt.hist(x, bins=20)
```

```python
n, bins, patches
```

```python
# density = True
n, bins, patches = plt.hist(x, bins=20, density=True)
```

```python
n, bins, patches = plt.hist(x, bins=20, density=True, facecolor='blue', alpha=0.5)
```

```python
# Setting X, Y label and Title, axis
n, bins, patches = plt.hist(x, bins=20, density=True, facecolor='blue', alpha=0.5)
plt.xlabel('X bins')
plt.ylabel('Percentage of observations')
plt.title('Histogram of Gaussian Distribution')
plt.axis([-10, 10, 0, 0.16])
plt.legend(title = 'random # from gauss dist')
plt.show()
```

### Kernel Density Curve

```python
# Kernel Density Curve with histogram
sns.distplot(x, 
            hist=True, 
            kde=True, 
            bins=20, 
            color='blue', 
            hist_kws={'edgecolor': 'gray'}, 
            kde_kws={'linewidth': 2})
```

```python
# Kernel Density Curve without histogram
sns.distplot(x, 
            hist=False, 
            kde=True, 
            bins=20, 
            color='blue', 
            hist_kws={'edgecolor': 'gray'}, 
            kde_kws={'linewidth': 2})
```

### Histogram and Density Plot with Multiple Groups

```python
# DataFrame with multiple groups
df = pd.DataFrame({'grp': ['grp1']*50 + ['grp2']*50 + ['grp3']*50, 
                  'x': np.random.randn(150)})
```

```python
# list of group
grp = ['grp1', 'grp2', 'grp3']

for group in grp:
    
    # subset to the group
    subset = df[df['grp'] == group]
    
    # kernel density curve
    sns.distplot(subset['x'], 
                hist = False, 
                kde = True, 
                kde_kws = {'linewidth': 2}, 
                label = group)

# setting plot format
plt.title('Kernel density curve by Group')
plt.xlabel('X')
plt.ylabel('density')
plt.legend(prop={'size': 12}, title = 'Group')
```

### Histogram with multiple variables

```python
import matplotlib.pyplot as plt
import seaborn as sns

iris = sns.load_dataset('iris')
iris.shape
```

```python
iris.head()
```

```python
# 1. Multiple histograms on the same axis
var_color_dict = {'sepal_length': 'blue', 
                 'sepal_width': 'red', 
                 'petal_length': 'yellow', 
                 'petal_width': 'green'}

for var in var_color_dict:
    sns.distplot(iris[var],                  
                 color=var_color_dict[var], 
                 hist_kws={'edgecolor': 'gray'}, 
                 label=var)

plt.legend(title='Species')
plt.show()
```

```python
# 2. Multiple histograms at separate windows
f, axes = plt.subplots(2, 2, figsize=(8, 6), sharex=True)
sns.distplot(iris["sepal_length"], color="blue", ax=axes[0, 0])
sns.distplot(iris["sepal_width"], color="red", ax=axes[0, 1])
sns.distplot(iris["petal_length"], color="yellow", ax=axes[1, 0])
sns.distplot(iris["petal_width"], color="green", ax=axes[1, 1])
plt.show()
```
